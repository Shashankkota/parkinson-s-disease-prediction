3)Extract the features and labels from the dataset. The features include all columns except â€˜statusâ€™, while the labels correspond to the values in the â€˜statusâ€™ column.

```python
#DataFlair - Get the features and labels
features=df.loc[:,df.columns!='status'].values[:,1:]
labels=df.loc[:,'status'].values
```

5)initialize a `MinMaxScaler` to normalize the features within a range of -1 to 1. This scaler adjusts the feature values proportionally within the specified range. The `fit_transform()` method is used to first compute the necessary scaling parameters and then apply the transformation. The labels remain unchanged as they do not require scaling.

```python
scaler=MinMaxScaler((-1,1))
x=scaler.fit_transform(features)
y=labels
```

6)Next, divide the dataset into training and testing sets, allocating 20% of the data for testing. This ensures the model is trained on the majority of the data while retaining a portion for evaluation.

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=7)
```

7)Create an instance of XGBClassifier and train the model. This classifier utilizes eXtreme Gradient Boosting, a powerful gradient boosting algorithm designed for modern data science challenges. It belongs to the Ensemble Learning category in machine learning, where multiple models are trained and combined to generate a more accurate and robust prediction.

```python
model=XGBClassifier()
model.fit(x_train,y_train)
```

Generate `y_pred`, which contains the predicted values for `x_test`, and evaluate the modelâ€™s accuracy. Finally, print the accuracy score to assess the model's performance.

```python
y_pred=model.predict(x_test)
print(accuracy_score(y_test, y_pred)*100)
```

In the given Parkinsonâ€™s Disease Detection Project, the model is trained on a dataset, but there is no direct user input to predict whether a patient has the disease.

Now we need to add User Input for Prediction
To allow user input (e.g., from a form or console), you need to:

Collect the required features from the user.
Preprocess the input like the training data.
Use the trained XGBClassifier model to predict if the patient has Parkinsonâ€™s disease.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv('parkinsons.data')

X = df.drop(columns=['name', 'status'])
y = df['status']

scaler = MinMaxScaler((-1, 1))
X_scaled = scaler.fit_transform(X)

x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(x_train, y_train)

def predict_parkinsons():
    print("Enter the following 22 values:")

    user_input = []
    feature_names = df.columns.drop(['name', 'status'])

    for feature in feature_names:
        value = float(input(f"{feature}: "))
        user_input.append(value)

    user_input = np.array(user_input).reshape(1, -1)
    user_input = scaler.transform(user_input)
    prediction = model.predict(user_input)

    if prediction[0] == 1:
        print("The model predicts that the patient has Parkinsonâ€™s Disease.")
    else:
        print("The model predicts that the patient does NOT have Parkinsonâ€™s Disease.")

predict_parkinsons()
```
